```python
import numpy as np
import pandas as pd
# matplotlib = Basic Plots made easy
import matplotlib.pyplot as plt
# Descriptive Plots made easy
import seaborn as sns
```

```python
auto_price = pd.read_csv("/Users/ahmetokur/Desktop/Datasets/Automobile_price_data__Raw_.csv", sep = ",")  # we can use URL here as an alternative
```

```python
auto_price
```

```python
# convert object columns to numerical columns
cols = ['bore','stroke','horsepower','peak-rpm','price']
auto_price[cols] = auto_price[cols].apply(pd.to_numeric, args = ('coerce',))
```

```python
auto_price.isnull().sum()
```

```python
auto_price.dtypes
```

```python
# Remove/Drop the Missing Value Rows
auto_price.dropna(inplace = True)
```

#### Line Plot
    
        * Relationship Plot (Bivariate)
        * Trends/Values - Univariate 

```python
# Using Pandas for Visualization
auto_price['price'].plot(figsize = (12,5))
```

```python
# alpha = for transparency
sns.set_style('whitegrid')
plt.figure(figsize = (16, 6))
plt.plot(auto_price['price'],  color = 'red', alpha = 0.8)
plt.title('Price of Vehicle Line Plot', fontsize = 14)
plt.xlabel('Row Indexes', fontsize = 10)
plt.ylabel('Price', fontsize = 10)
plt.show()
```

### Bar Plot
* Bivariate Analysis
*Bar chart used by Catogorical, Nominal, Catogorical vs Numerical/Continious
* Bar plots are used ti display counts of uniq value of catagorical dtype, 
Height of bar represents counts for eachcatagory

```python
sns.set_style('whitegrid')
auto_price[['price']][10:70].plot.bar(figsize = (15,7))
```

```python
counts = auto_price['make'].value_counts()
counts
```

```python
# how many car produce every maker
fig = plt.figure(figsize = (7,6))
auto_price['make'].value_counts().plot.bar(figsize = (15,7))
plt.show()
```

```python
# for every row we are getting the data
auto_price[['city-mpg','horsepower','wheel-base']][10:60].plot.bar(figsize = (20,8), width = 0.8)
```

### Histogram
* Univariate values

```python
fig = plt.figure(figsize = (10,6))
# bins = intervals
plt.hist(auto_price['price'], color = 'lightgreen', bins = 15)
plt.title('Histogram Plot - Auto Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

### Subplots


```python
# Subplots = (1, 2, 1) => 1st Row, 2nd Column, 1st Index
plt.figure(figsize = (18,6))
plt.subplot(1, 2, 1)
plt.hist(auto_price['engine-size'], color = 'deepskyblue', bins = 20)
plt.title('Histogram Plot - Engine Size')
plt.xlabel('Engine Size')
plt.ylabel('Number of Vehicle')
plt.subplot(1, 2, 2)
plt.hist(auto_price['price'], color = 'lightgreen', bins = 20)
plt.title('Histogram Plot - Auto Price')
plt.xlabel('Price')
plt.ylabel('Number of Vehicle')
plt.show()
```

#### Scatter Plot
    * Scatter Plot is plotted between numerical values. 
    * Shows relationship b/w x & y plots

```python
plt.figure(figsize = (10,7))
# marker = changing the scatter plots style o,x,X,s,d,>,<,*,.,^
sns.scatterplot(x = auto_price['city-mpg'], y = auto_price['price'], color = 'crimson', marker = 'o')
plt.show()
```

```python
plt.figure(figsize = (10,7))
sns.scatterplot(x = auto_price['city-mpg'], y = auto_price['price'], hue = auto_price['fuel-type'], color = 'crimson')
plt.show()
```

### Box Plot
* Distrubution of sample data
* Five point Summary - min, max, Q1, median, Q3 
* Outliers

```python
plt.figure(figsize = (10,5))
sns.boxplot(x = 'price', data = auto_price)
```

```python
# https://matplotlib.org/stable/tutorials/colors/colormaps.html for cmap!
plt.figure(figsize = (12,8))
sns.boxplot(x = 'drive-wheels', y = 'price', data = auto_price, palette = 'Dark2')
```

### Distrubution Plot
* Histogram + Density
* Probability density function - Continious

```python
plt.figure(figsize = (10,6))
# hist = False : Disable Histogram
sns.distplot(auto_price['price'], color = 'crimson', hist = False)
plt.show()
```

### Pie-Chart

```python
plt.figure(figsize = (10,10))
# Shows distribution % in upto 2 decimal Places
plt.pie(auto_price['make'].value_counts(), autopct = '%0.2f%%')
plt.show()
```

```python
auto_price['make'].value_counts().index
```

```python
plt.figure(figsize = (10,10))
# Shows distribution % in upto 2 decimal Places
plt.pie(auto_price['make'].value_counts(), autopct = '%0.2f%%', labels = auto_price['make'].value_counts().index)
plt.show()
```

### HeatMap
* Multivariate data

```python
auto_price.corr()
# corr = 0 or closer to 0
# if b/w two numerical values corr is +ve it means both are directly proprtional, 
# -ve means both are inversly proprtional.
```

```python
plt.figure(figsize = (2,9))
sns.heatmap(auto_price.corr()[['price']], annot = True, cmap = 'magma')
plt.show()
# Stronger correlation is closer to +1 (light color)
```

```python
plt.figure(figsize = (10,9))
sns.heatmap(auto_price.corr(), annot = True, cmap = 'magma')
plt.show()
```
